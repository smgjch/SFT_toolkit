```python
import os
import json

def remove_null_content_from_json(json_data):
    chats_to_save = []
    null_cnt = 0
    len_cnt = 0
    for item in json_data:

        if check_structure(item):
            if item["messages"][0]["content"] != None and item["messages"][1]["content"] != None:
                chats_to_save.append(item)
            else:
                null_cnt+=1
        else:
            print(item)
            len_cnt += 1
    print(f"found {null_cnt} null")
    print(f"found {len_cnt} illegal chat")

    return chats_to_save

def check_structure(item):
    return "messages" in item and len(item["messages"]) == 2 and item["messages"][0]["role"] == "user" and item["messages"][1]["role"] == "assistant"

def process_json_files_in_directory(directory_path):
    for filename in os.listdir(directory_path):
        if filename.endswith('.json'):
            file_path = os.path.join(directory_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                try:
                    data = json.load(file)
                except json.JSONDecodeError:
                    print(f"Error decoding JSON in file: {filename}")
                    continue

            cleaned_data = remove_null_content_from_json(data)

            with open(file_path, 'w', encoding='utf-8') as file:
                json.dump(cleaned_data, file, indent=4,ensure_ascii=False)

            print(f"Processed file: {filename}")

# Example usage
directory_path = '/mnt/petrelfs/hujucheng/SFT_tools/SFT_toolkit/data/1128full'
process_json_files_in_directory(directory_path)
```
