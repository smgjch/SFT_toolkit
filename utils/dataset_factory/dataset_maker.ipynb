```python
def label_last_five_paragraphs(input_string):
    # Split the input string into paragraphs
    paragraphs = input_string.split('\n')
    
    # Check if there are at least five paragraphs
    num_paragraphs = len(paragraphs)
    start_index = max(0, num_paragraphs - 6)  # Calculate the starting index for the last five
    
    # Labels for the options
    labels = ['A.', 'B.', 'C.', 'D.', 'E.']
    
    # Modify the last five paragraphs
    for i in range(start_index, num_paragraphs):
        label_index = i - start_index  # Calculate the label index
        if label_index < len(labels):  # Check if we have enough labels
            paragraphs[i] = labels[label_index] + ' ' + paragraphs[i]
    
    # Join the paragraphs back into a single string
    output_string = '\n'.join(paragraphs)
    return output_string
```

```python
def compress_and_replace_tabs_label(input_string):
    input_string = label_last_five_paragraphs(input_string)
    # Replace all tabs with spaces
    no_tabs_string = input_string.replace('\t', ' ')
    
    # Split the string into lines
    lines = no_tabs_string.splitlines()
    
    # Join the lines using \n as the separator
    compressed_string = '\\n'.join(lines)
    
    return compressed_string
def compress_and_replace_tabs(input_string):
    # input_string = label_last_five_paragraphs(input_string)
    # Replace all tabs with spaces
    no_tabs_string = input_string.replace('\t', ' ')
    
    # Split the string into lines
    lines = no_tabs_string.splitlines()
    
    # Join the lines using \n as the separator
    compressed_string = '\\n'.join(lines)
    
    return compressed_string
```

```python
def parse_md_to_json(md_content):
    # Split the content by lines
    lines = md_content.split('\n')
    
    # Initialize a list to hold each parsed question block
    parsed_data = []
    current_block = None
    current_role = None
    
    for line in lines:
        if line.startswith('# ID:'):
            if current_block:
                parsed_data.append(current_block)
            current_block = {'messages': []}
            current_role = 'user'
        elif line.startswith('Correct Answer:'):
            current_role = 'assistant'
            # Ensure there is already a user message to modify
            if current_block and current_block['messages']:
                current_block['messages'].append({
                    "role": 'assistant',
                    "content": ""
                })
        else:
            if current_role and current_block:
                # Add content to the current role's last message, or create new one if needed
                if current_block['messages'] and current_block['messages'][-1]['role'] == current_role:
                    current_block['messages'][-1]['content'] += line + '\n'
                else:
                    current_block['messages'].append({
                        "role": current_role,
                        "content": line + '\n'
                    })
    
    # Append the last block if it exists
    if current_block:
        parsed_data.append(current_block)

    return parsed_data
```

```python
def parse_txt_to_json(txt_content):
    # Split the content into blocks by double newlines, assuming each block is a separate question
    blocks = txt_content.strip().split('\n\n')
    parsed_data = []

    for block in blocks:
        lines = block.strip().split('\n')
        if lines:
            # The main question text is the first part
            question_text = ' '.join(lines[:-4])  # Assuming the last four lines are the options
            options = lines[-4:]  # The last four lines are the options

            # Assume the correct answer is always the first option
            correct_answer = question_text[0]  # 'A', 'B', 'C', 'D'

            # Construct the messages
            question_block = {
                "messages": [
                    {
                        "role": "user",
                        "content": question_text[1:] + " " + " ".join(options)
                    },
                    {
                        "role": "assistant",
                        "content": f"{correct_answer.upper()}"
                    }
                ]
            }
            parsed_data.append(question_block)

    return parsed_data
```

```python
import json
import random
def slice_json(input_json, num_slices=53):
    # Open the input JSON file
    with open(input_json, 'r') as file:
        data = json.load(file)
    
    random.shuffle(data)
    # Determine the slice size
    total_items = len(data)
    slice_size = total_items // num_slices
    remainder = total_items % num_slices

    # Create and save each slice
    start = 0
    for i in range(num_slices):
        end = start + slice_size + (1 if i < remainder else 0)
        slice_data = data[start:end]
        
        # Write each slice to a separate file
        with open(f'/mnt/petrelfs/hujucheng/train/data/10Wbase/70Wbase_sliced/70wbase_{i+1}.json', 'w') as outfile:
            json.dump(slice_data, outfile,  indent=4,ensure_ascii=False)
        
        # Update start for the next slice
        start = end

    print(f"Data has been sliced into {num_slices} sections and saved to files.")

# Example usage
input_json = '/mnt/petrelfs/hujucheng/train/data/10Wbase/70Wbase/70wBase.json' 

slice_json(input_json)
```

```python
import json
import re
import re

def parse_question_and_answer(text):
    question_pattern = re.compile(r"<question>(.*?)<question>", re.DOTALL)
    answer_pattern = re.compile(r"<Answer>(.*?)<Answer>", re.DOTALL)

    question_match = question_pattern.search(text)
    answer_match = answer_pattern.search(text)

    question = question_match.group(1).strip() if question_match else None
    answer = answer_match.group(1).strip() if answer_match else None

    return question, answer

def parsing_translation(LLM_answer, output_file):
    # Read the input JSON from a file
    with open(LLM_answer, "r") as infile:
        LLM_answer_json = json.load(infile)


    answer_set_json = LLM_answer_json["details"]
    output_json = []
    for key, value in answer_set_json.items():
        response = value["origin_prediction"]
        # print(response,"\n----------------------------------")

        question, answer = parse_question_and_answer(response)
        if question != None and answer != None:
            message = {
                "messages": [
                    {
                        "role": "user",
                        "content": question
                    },
                    {
                        "role": "assistant",
                        "content": answer
                    }
                ]
            }
            output_json.append(message)
    
    # Write the transformed JSON to the output file
    with open(output_file, "w") as outfile:
        json.dump(output_json, outfile, indent=4, ensure_ascii=False)

    print(f"Explianed JSON has been saved to {output_file}")
```

```python
import json

def cast_dart_to_openai_format(input_file1, input_file2,output_file):
    # Read the input JSON from a file
    with open(input_file1, "r") as infile:
        input_json1 = json.load(infile)  

    with open(input_file2, "r") as infile:
        input_json2 = json.load(infile)
    
    output_json = []

    querys = input_json1["query"]
    respeonese = input_json1["response"]    
    querys2 = input_json2["query"]
    respeonese2 = input_json2["response"]


    for idx in querys:
       
        question = querys[idx]
        answer_text = respeonese[idx]
        
        message = {
            "messages": [
                {
                    "role": "user",
                    "content": question
                },
                {
                    "role": "assistant",
                    "content": answer_text
                }
            ]
        }
        output_json.append(message)
    
        question2 = querys2[idx]
        answer_text2 = respeonese2[idx]
        message2 = {
            "messages": [
                {
                    "role": "user",
                    "content": question2
                },
                {
                    "role": "assistant",
                    "content": answer_text2
                }
            ]
        }
        output_json.append(message2)

    # Write the transformed JSON to the output file
    with open(output_file, "w") as outfile:
        json.dump(output_json, outfile, indent=4,ensure_ascii=False)

    print(f"Dart Set JSON has been saved to {output_file}")
```

```python
import os
import json
import random

def sample_json_items(directory, sample_percentage=1/7):
    all_items = []

    # Iterate over all files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.json'):
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r') as file:
                data = json.load(file)
                
                # Assuming each file contains a list of items
                if isinstance(data, list):
                    all_items.extend(data)
                else:
                    print(f"Warning: {filename} does not contain a list.")

    # Calculate the number of items to sample
    sample_size = int(len(all_items) * sample_percentage)
    
    # Randomly sample the items
    sampled_items = random.sample(all_items, sample_size)
    
    return sampled_items

# Usage
directory_path = "/mnt/hwfile/opendatalab/peiqizhi/data/reduced_ds_v1_data_70w"
sampled_items = sample_json_items(directory_path)

```

```python
import json

def transform_LSAT_AR(input_file, output_file):
    # Read the input JSON from a file
    with open(input_file, "r") as infile:
        input_json = json.load(infile)
    
    output_json = []
    background = input_json["Setup"]
    questions = input_json["QuestionWithOptions"]
    answers = input_json["Correct Answer"]
    for idx in background:

        f_string = f"{background[idx]}\n{questions[idx][2:]}"
        answer = answers[idx]
        # Construct the message template
        message = {
            "messages": [
                {
                    "role": "user",
                    "content": f_string
                },
                {
                    "role": "assistant",
                    "content": answer
                }
            ]
        }
        output_json.append(message)
    
    # Write the transformed JSON to the output file
    with open(output_file, "w") as outfile:
        json.dump(output_json, outfile, indent=4,ensure_ascii=False)

    print(f"Transformed JSON has been saved to {output_file}")

# Example usage
input_file_path = "/mnt/petrelfs/hujucheng/SFT_tools/SFT_toolkit/data/analytical_reasoning/from_hugging_face/LAST-ar_addition_from_huggingface_raw.json"  # Replace with your input file path
output_file_path = "/mnt/petrelfs/hujucheng/SFT_tools/SFT_toolkit/data/analytical_reasoning/from_hugging_face/LAST-ar_addition_from_huggingface.json"  # Replace with your output file path

transform_LSAT_AR(input_file_path, output_file_path)
```

```python
import json

def transform_LSAT_AR_github(input_file, output_file):
    # Read the input JSON from a file
    with open(input_file, "r") as infile:
        input_json = json.load(infile)
    
    output_json = []

    for item in input_json:
        background = item["passage"]
        for question in item["questions"]:
            question_text = f"{background}\n{question["question"]}\n"
            choices = question["options"]
            options = []
            for idx, choice in enumerate(choices):
                letter = chr(65 + idx )  # 'A', 'B', 'C', etc.
                option = f"{letter}. {choice}\n"
                options.append(option)
            question_text += "".join(options)

            answer = f"{item["passageBreakdown"]} {question["explanation"]}"

            message = {
                "messages": [
                    {
                        "role": "user",
                        "content": question_text
                    },
                    {
                        "role": "assistant",
                        "content": answer
                    }
                ]
            }
            output_json.append(message)
    
    # Write the transformed JSON to the output file
    with open(output_file, "w") as outfile:
        json.dump(output_json, outfile, indent=4,ensure_ascii=False)

    print(f"Transformed JSON has been saved to {output_file}")

# Example usage
input_file_path = "/mnt/petrelfs/hujucheng/SFT_tools/SFT_toolkit/data/analytical_reasoning/from_gitHub/AR_DevelopmentDataExplained.json"  # Replace with your input file path
output_file_path = "/mnt/petrelfs/hujucheng/SFT_tools/SFT_toolkit/data/analytical_reasoning/from_gitHub/LAST-ar_addition_from_gitHub.json"  # Replace with your output file path

transform_LSAT_AR_github(input_file_path, output_file_path)
```
